```python
import pandas as pd
import numpy as np
import altair as alt
import matplotlib.pyplot as plt
import ppscore as ps
import seaborn as sns
```

```python
df = pd.read_csv('Haircut_Tip_Amounts.csv.xls')
```

```python
# Correlation matrix
df = pd.read_csv('Haircut_Tip_Amounts.csv.xls')
df['Tip amount'] = df['Tip amount'].replace({
    '3,5': 3.5,
    '4,5':4.5
})
df['Tip amount']=df['Tip amount'].astype(int)

df['Married'] = df['Married'].replace({
    'Kid': '1',
    'M': 0,
    'NM': 1
})
df['Married']=df['Married'].astype(int)

df['Day'] = df['Day'].replace({
    'mon': 0,
    'Tues': 1,
    'Thurs': 2,
    'Mon': 0,
    'Tue': 1,
    'Wed': 2,
    'Thur': 3,
    'Fri': 4,
    'Sat': 5,
    'Sun': 6
})
df['Day']=df['Day'].astype(int)

df['Culture'] = df['Culture'].replace({
    'I': 0,
    'E': 1,
    'B': 2,
    'W': 3,
    'P': 4,
    'M': 5,
    'A': 6,
    'H': 7,
    'V':8
})
df['Culture']=df['Culture'].astype(int)
df.corr(method='spearman')
```

```python
# Heatmap of correlation matrix
sns.heatmap(df.corr(method='spearman'),annot=True)
```

```python
# Tip on Age grouped
df = pd.read_csv('Haircut_Tip_Amounts.csv.xls')
df['Tip amount'] = df['Tip amount'].replace({
    '3,5': 3.5,
    '4,5': 4.5
})
bins= [0,15,30,45,60,85]
labels = ['0-14','15-29','30-44','45-59','60+']

df['Tip amount']=df['Tip amount'].astype(int)
df.groupby(pd.cut(df['Age'], bins=bins, labels=labels, right=False)).agg({'Age': ['size'], 'Tip amount': ['mean']})
```

```python
# Tip on Married and Non-Married
df = pd.read_csv('Haircut_Tip_Amounts.csv.xls')
df['Married'] = df['Married'].replace({
    'Kid': 'NM'
})

df['Tip amount'] = df['Tip amount'].replace({
    '3,5': 4,
    '4,5': 5
})

df['Tip amount']=df['Tip amount'].astype(int)
df = df.groupby('Married', as_index=False)[['Tip amount']].mean()

alt.Chart(df).mark_bar(size=35).encode(
    x='Married',
    y='Tip amount',
    tooltip=['Tip amount']).properties(width=200)
```

```python
# Tip on Days
df = pd.read_csv('Haircut_Tip_Amounts.csv.xls')
df['Day'] = df['Day'].replace({
    'mon': 'Mon',
    'Tues': 'Tue',
    'Thurs': 'Thur',
})

df['Tip amount'] = df['Tip amount'].replace({
    '3,5': 4,
    '4,5': 5
})

df['Tip amount']=df['Tip amount'].astype(int)
df = df.groupby('Day', as_index=False)[['Tip amount']].mean()

alt.Chart(df).mark_bar(size=35).encode(
    x='Day',
    y='Tip amount',
    tooltip=['Tip amount']).properties(width=600)
```

```python
# Tip on Cultrue
df = pd.read_csv('Haircut_Tip_Amounts.csv.xls')
df['Tip amount'] = df['Tip amount'].replace({
    '3,5': 4,
    '4,5': 5
})

df['Tip amount']=df['Tip amount'].astype(int)
df = df.groupby('Culture', as_index=False)[['Tip amount']].mean()

alt.Chart(df).mark_bar(size=35).encode(
    x='Culture',
    y='Tip amount',
    tooltip=['Tip amount']).properties(width=600)
```

```python
# Tip on Time
df = pd.read_csv('Haircut_Tip_Amounts.csv.xls')
df['Tip amount'] = df['Tip amount'].replace({
    '3,5': 4,
    '4,5': 5
})

df['Time'] = df['Time'].replace({
    1:13,
    2:14,
    3:15,
    4:16,
    5:17,
    6:18,
    7:19,
    8:20,
})


df['Tip amount']=df['Tip amount'].astype(int)
df = df.groupby('Time', as_index=False)[['Tip amount']].mean()

alt.Chart(df).mark_bar(size=15).encode(
    x='Time',
    y='Tip amount',
    tooltip=['Tip amount','Time']).properties(width=500)
```
